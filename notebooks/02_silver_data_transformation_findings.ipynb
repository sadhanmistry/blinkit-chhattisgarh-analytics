```python
# actual_delivery_time_minutes is null for a lot of rows -- before treating
# this as dirty data, check whether it lines up with cancelled orders, where a
# missing delivery time is actually the CORRECT/expected value.
cancelled_ids = set(orders.loc[orders['order_status'] == 'Cancelled', 'order_id'])
null_time_mask = deliveries['actual_delivery_time_minutes'].isnull()
in_cancelled = deliveries.loc[null_time_mask, 'order_id'].isin(cancelled_ids)
print("null actual_delivery_time_minutes for CANCELLED orders (expected, not dirty):", in_cancelled.sum())
print("null actual_delivery_time_minutes for NON-cancelled orders (genuinely missing):", (~in_cancelled).sum())
```

```python
deliveries['actual_delivery_time_minutes'].describe()
```

```python
# anything above ~100 minutes is operationally impossible for quick commerce
(deliveries['actual_delivery_time_minutes'] > 100).sum()
```

```python
deliveries.loc[deliveries['actual_delivery_time_minutes'] > 100, 'actual_delivery_time_minutes'].describe()
```

```python
# Findings
# 1:- Null Values:-
                #1:- actual_delivery_time_minutes -- the majority of these nulls
                #     line up exactly with Cancelled orders (correct, not dirty).
                #     A smaller number do not, and ARE genuinely missing data.
                #2:- distance_km -- genuinely missing, impute with store median
                #3:- rider_id -- genuinely missing for a small number of rows;
                #     can't be reliably derived, so flag rather than fabricate

# 2:- Outliers:-
                # actual_delivery_time_minutes has values as high as ~400 minutes
                # for NON-cancelled orders -- clearly a data error, not a real
                # slow delivery. Use a fixed threshold (100 min), not a
                # percentile cutoff, since percentile capping would also clip
                # genuinely slow (but real) deliveries in the long tail.
```
