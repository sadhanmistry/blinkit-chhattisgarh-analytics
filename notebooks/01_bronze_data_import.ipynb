```python
from sqlalchemy import create_engine

engine = create_engine(
    "mysql+pymysql://root:password@localhost/blinkit_bronze"
)

files = {
    "stores":        "/Users/sadhanmistry/Desktop/blinkit/blinkit_cg_dataset/stores.csv",
    "customers":     "/Users/sadhanmistry/Desktop/blinkit/blinkit_cg_dataset/customers.csv",
    "products":      "/Users/sadhanmistry/Desktop/blinkit/blinkit_cg_dataset/products.csv",
    "riders":        "/Users/sadhanmistry/Desktop/blinkit/blinkit_cg_dataset/riders.csv",
    "orders":        "/Users/sadhanmistry/Desktop/blinkit/blinkit_cg_dataset/orders.csv",
    "order_items":   "/Users/sadhanmistry/Desktop/blinkit/blinkit_cg_dataset/order_items.csv",
    "deliveries":    "/Users/sadhanmistry/Desktop/blinkit/blinkit_cg_dataset/deliveries.csv",
    "inventory":     "/Users/sadhanmistry/Desktop/blinkit/blinkit_cg_dataset/inventory.csv"
}

for table_name, file_name in files.items():
    df = pd.read_csv(file_name)
    df.to_sql(table_name, con=engine, if_exists="replace", index=False)
    print(f"Loaded {table_name}: {len(df):,} rows")
```
