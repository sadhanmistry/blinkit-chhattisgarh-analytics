```python
import pandas as pd
from sqlalchemy import create_engine
import os

gold_engine = create_engine("mysql+pymysql://root:password@localhost/blinkit_gold")
OUT_DIR = "blinkit_gold_csv"
os.makedirs(OUT_DIR, exist_ok=True)

TABLES = ["dim_date", "dim_stores", "dim_customers", "dim_products", "dim_riders",
          "fact_orders", "fact_order_items", "fact_deliveries", "fact_inventory"]

for name in TABLES:
    df = pd.read_sql(f"SELECT * FROM {name}", gold_engine)
    df.to_csv(f"{OUT_DIR}/{name}.csv", index=False)
    print(f"exported {name}: {len(df):,} rows")
```
